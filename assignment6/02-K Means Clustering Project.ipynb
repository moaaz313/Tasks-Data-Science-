## K Means Cluster Creation

Now it is time to create the Cluster labels!

** Import KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

** Create an instance of a K Means model with 2 clusters.**

```python
# Create an instance of KMeans with 2 clusters
kmeans = KMeans(n_clusters=2)
```

**Fit the model to all the data except for the Private label.**

```python
X = df.drop(columns=['Private', 'Unnamed: 0'])  # Remove 'Private' and 'Name' columns

kmeans.fit(X)  # Fit the KMeans model to the data
```

** What are the cluster center vectors?**

```python
# Get the cluster center vectors
cluster_centers = kmeans.cluster_centers_

# Display the cluster center vectors
print(cluster_centers)
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters, keep in mind, you usually won't have this luxury in the real world.

** Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
# Create the 'Cluster' column based on the 'Private' column
df['Cluster'] = df['Private'].apply(lambda x: 1 if x == 'Yes' else 0)

# Display the first few rows to check the new column
df[['Unnamed: 0', 'Private', 'Cluster']].head()
```

** Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.**

```python
# Get the predicted cluster labels from the KMeans model
predicted_labels = kmeans.labels_
```

```python
# Compare the number of private schools (1) in each cluster
private_in_cluster_0 = sum((predicted_labels == 0) & (df['Cluster'] == 1))
private_in_cluster_1 = sum((predicted_labels == 1) & (df['Cluster'] == 1))

# If cluster 0 has more private schools, swap the labels
if private_in_cluster_0 > private_in_cluster_1:
    predicted_labels = 1 - predicted_labels  # Swap the labels: 0 becomes 1, 1 becomes 0
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Generate the confusion matrix
conf_matrix = confusion_matrix(df['Cluster'], predicted_labels)

# Generate the classification report
class_report = classification_report(df['Cluster'], predicted_labels)

# Display the confusion matrix and classification report
print("Confusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```

Not so bad considering the algorithm is purely using the features to cluster the universities into 2 distinct groups! Hopefully you can begin to see how K Means is useful for clustering un-labeled data!
